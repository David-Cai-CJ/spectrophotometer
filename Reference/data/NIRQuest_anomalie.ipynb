```python
import numpy as np
import matplotlib.pylab as plt

plt.style.use(['fast','ggplot'])

%config InlineBackend.figure_format='retina'
```

```python
x = np.load('wavelengths.npy')
ms_1 = np.load('1ms.npy')
ms_100 = np.load('100 ms.npy')
```

```python

plt.plot(x, ms_1[0])
plt.title('Intensity over 1 ms (Unexposed)')

# plt.yscale('log')
plt.ylabel("Counts")
plt.xlabel("Wavelengths (nm)")
# plt.axvline(1319, color ='k', ls = '--', alpha = .5, label = '1319 nm')
# plt.axvline(2400, color ='k', ls = '--', alpha = .5, label = '2400 nm')
plt.savefig('1ms_signal.jpg', dpi =300, bbox_inches ='tight')
```

```python
plt.plot(x, np.std(ms_1, axis = 0 ), label = 'Measured STD')
plt.plot(x, np.sqrt(ms_1[2]), label = 'Poissonian')
plt.title('Intensity flucutations\nover 1 ms Intervals')

# plt.yscale('log')
plt.ylabel("$\Delta$ Counts")
plt.xlabel("Wavelengths (nm)")
plt.axvline(1319, color ='k', ls = '--', alpha = .5, label = '1319 nm')
plt.axvline(2400, color ='k', ls = '--', alpha = .5, label = '2400 nm')
plt.legend()
plt.savefig('1ms_fluc.jpg', dpi =300, bbox_inches ='tight')
```

```python
plt.plot(x, ms_100[0])
plt.title('Intensity over 100 ms (Unexposed)')

plt.ylabel("Counts")
plt.xlabel("Wavelengths (nm)")
plt.savefig('100ms_signal.jpg', dpi =300, bbox_inches ='tight')
```

```python
plt.plot(x, np.std(ms_100, axis = 0 ), label = 'Measured STD')
plt.plot(x, np.sqrt(ms_100[2]), label = 'Poissonian')
plt.title('Intensity flucutations\nover 1 ms Intervals')

# plt.yscale('log')
plt.ylabel("$\Delta$ Counts")
plt.xlabel("Wavelengths (nm)")
plt.axvline(1319, color ='k', ls = '--', alpha = .5, label = '1319 nm')
plt.axvline(2400, color ='k', ls = '--', alpha = .5, label = '2400 nm')
plt.legend()
plt.savefig('100ms_fluc.jpg', dpi =300, bbox_inches ='tight')
```

```python
np.std(ms_1, axis = 0).mean()
```

```python
plt.plot(np.sqrt(np.std(ms_1, axis = 0) **2 *100), label = 'STD 1 ms Scaled to 100 ms')
plt.plot(np.std(ms_100, axis = 0), label = 'STD 100 ms', zorder= 3)
plt.yscale('log')
plt.legend()
```

```python
no_lamp = np.load('no_lamp_1ms.npy')
lamp = np.load('lamp_1ms.npy')
```

```python
plt.plot(x, lamp)
plt.plot(x, no_lamp)
```

```python
elamp = np.sqrt(lamp)
eno_lamp = np.sqrt(no_lamp)
eratio = np.sqrt((elamp/no_lamp)**2 + (lamp/no_lamp**2*eno_lamp)**2)
```

```python
plt.errorbar(x, lamp/no_lamp, yerr= eratio, elinewidth = 1,capsize = 1, marker= 'None',ls = 'None',
            label='Tungsten\nHalogen Lamp')
plt.xlabel("Wavelength (nm)")
plt.ylabel('SNR')
plt.legend(frameon = True, framealpha = .3, loc = 1, facecolor='gray')
plt.savefig('NIRQuest_dead_pixels.pdf', bbox_inches= 'tight')
```
